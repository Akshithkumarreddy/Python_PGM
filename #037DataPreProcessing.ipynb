```python
# Taking care of missing data  Through sklearn
#from sklearn.preprocessing import Imputer
from sklearn.impute import SimpleImputer
#imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = SimpleImputer(strategy='mean')
s1 = imputer.fit_transform(x[['Age','Salary']])
#x.loc[:,['Age','Salary']] = imputer.fit_transform(x[['Age','Salary']])
#imputer = SimpleImputer(strategy="most_frequent")
imputer = SimpleImputer(strategy='constant', fill_value='India')

#x.loc[:,'Country'] = imputer.fit_transform(x[['Country']])
s2 = imputer.fit_transform(x[['Country']])
```

```python
s2
```

```python
s1 = pd.DataFrame(s1,columns=['Age','Salary'])
s1['country'] = s2
```

```python
s1
```

```python
#Encode Categorical Data into Values using Pandas
s1['country'].replace('France','1')
s1['country'].replace('Spain','0')
s1['country'].replace('Germany','2')
#s1['country'].replace('India','4',inplace=True)
```

```python
s1
```

```python
s1['Country_encoded'] = s1.country.map({'France':1,'Spain':0,'Germany':2,'India':4})
```

```python
s1
```

```python
Countries = pd.get_dummies(s1.country)
```

```python
Countries
```

```python
s1 = pd.concat([s1,Countries],axis =1)
```

```python
s1
```

```python
dropcolums = ['country','France']
s1.drop(dropcolums,axis=1,inplace=True)
```
