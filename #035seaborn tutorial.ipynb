```python
#Facet Grid - 

tips_g = sns.FacetGrid(tips_df,col='time',row='smoker')
tips_g.map(plt.hist,'total_bill',bins=10)
```

```python
tips_g = sns.FacetGrid(tips_df,col='time',row='smoker')
tips_g.map(plt.scatter,'total_bill','tip')
```

```python
tips_g = sns.FacetGrid(tips_df,col='time',hue='smoker',height=5,aspect=1.5)
tips_g.map(plt.scatter,'total_bill','tip')
```

```python
tips_g = sns.FacetGrid(tips_df,col='time',hue='smoker',height=3,aspect=1.5,col_order=['Dinner','Lunch'])
tips_g.map(plt.scatter,'total_bill','tip',edgecolor='w')
```

```python
kws = dict(s=50,linewidth=.5,edgecolor='w')
tips_g = sns.FacetGrid(tips_df,col='sex',hue='smoker',height=4,aspect=1.4,hue_order=['Yes','No'],
                       hue_kws=dict(marker=['^','>']))
tips_g.map(plt.scatter,'total_bill','tip',**kws)
```
