```python
with open('people-1000.csv','r') as cust_file:
  csv_reader = csv.DictReader(cust_file)

  for line in csv_reader:
    print(line)
```

```python
with open('people-1000.csv','r') as cust_file:
  csv_reader = csv.DictReader(cust_file)

  for line in csv_reader:
    print(line['Email'])
```

```python
with open('people-1000.csv','r') as cust_file:
  csv_reader = csv.DictReader(cust_file)

  with open('new_people.csv','w') as new_file:
    fieldnames = ['First Name','Last Name','Sex','Email']
    csv_writer = csv.DictWriter(new_file,fieldnames=fieldnames,delimiter='\t')

    csv_writer.writeheader()

    for line in csv_reader:
       del line['User Id'] # below lines will not use these fields while creating new file
       del line['Phone']
       del line['Index']
       del line['Date of birth']
       del line['Job Title']
       csv_writer.writerow(line)
```
